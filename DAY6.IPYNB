```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datasets/titanic.csv")
df
```

```python
df.info()
df.isnull()
```

```python
age_imputation = df['Age'].fillna(df['Age'].mean())
print(age_imputation)
```

```python
print(df['Embarked'].mode().iloc[0])
print(df['Embarked'].info())
```

```python
df['Embarked'].fillna(df['Embarked'].mode().iloc[0])
```

```python
plt.subplot(2,3,1)
y = df['Pclass'].value_counts()
x = [1,2,3]
plt.bar(x,y)
plt.title('pclass - distribution')

plt.subplot(2,3,2)
plt.hist(df['Survived'] )
plt.title('Survival No')

plt.subplot(2,3,3)
plt.hist(df['Sex'])
plt.title('Gender Ratio')

plt.subplot(2,3,4)
plt.hist(df['Age'],bins=5)
plt.title('Age Distribution')

plt.subplot(2,3,5)
x = ['S','C','Q']
y=df['Embarked'].value_counts()
plt.bar(x,y)
plt.title('Embarked')
```

```python
df['Embarked'] = [1 if i=='S' else i for i in df['Embarked']]
df['Embarked'] = [2 if i=='C' else i for i in df['Embarked']]
df['Embarked'] = [3 if i=='Q' else i for i in df['Embarked']]
df['Embarked']
```

```python
df['Sex'] = [1 if i=='male' else i for i in df['Sex']]
df['Sex'] = [0 if i=='female' else i for i in df['Sex']]
df['Sex']
```

```python
columns = ['Survived','Pclass','Age','Sex','Embarked','Parch','SibSp']
correlation = df[columns].corr()
correlation
```

```python
sns.heatmap(correlation)
```

```python
sns.pairplot(df[columns],hue = 'Survived')
```

```python
sns.histplot(df['Age'])
```

```python
plt.figure(figsize=(15,8))
sns.kdeplot(df["Age"][df.Survived == 1], color="darkturquoise", shade=True)
sns.kdeplot(df["Age"][df.Survived == 0], color="lightcoral", shade=True)
plt.legend(['Survived', 'Died'])
plt.title('Density Plot of Age for Surviving Population and Deceased Population')
plt.show()
```
