```python
def f(x):
    return x**2-4*x+4

def diff(x):
    return 2*x-4
```

```python
class GD : 
    

    def __init__(self,lr,epochs):
        self.lr = lr 
        self.epochs = epochs
        self.weight = 100
        self.w = []

    def fit(self):
        for i in range(0,self.epochs):
            self.w.append(self.weight)
            self.weight = self.weight - self.lr*diff(self.weight)
        return self.w


```

```python
grad = GD(0.1,500)
```

```python
w = grad.fit()
```

#LINEAR REGRESSION 

```python
import sklearn.datasets as sd

X,y = sd.fetch_california_housing(return_X_y = True , as_frame = True)
```

```python
x = X.iloc[:,0:1]
x
```

```python
y
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, s=5)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
def reg(m,b,num):
    return m*num+b
```

```python
def diff(n,m,b,x_train,y_train):
    if n ==0 : 
        sum = 0
        for i in range(len(x_train)):
            sum += (reg(m,b,x_train.iloc[i])-y_train.iloc[i])
        
        return (-2*sum)
    elif n == 1 :
        sum = 0
        for i in range (len(x_train)):
            sum+= (reg(m,b,x_train.iloc[i])-y_train.iloc[i])*x_train.iloc[i]
        return (-2*sum)
```

```python
class grad_des:
    def __init__(self,lr,epoch):
        self.lr = lr
        self.epoch = epoch
        self.m = 100
        self.b = 1

        self.bias = []
        self.weight = []

    def fit(self,x_train,y_train):
        
        for i in range(0,self.epoch):
            self.bias.append(self.b)
            self.weight.append(self.m)
            self.b = self.b - self.lr*(diff(0,self.m,self.b,x_train,y_train))
            self.m = self.m - self.lr*(diff(1,self.m,self.b,x_train,y_train))
            
        return (self.weight,self.bias)

    def pred(self,x_test):
        y_pred = []
        for i in range(len(x_test)):
            y_pred.append(self.weight[-1]*x_test[i]+self.bias[-1])

        return y_pred

```

```python
from sklearn.model_selection import train_test_split     
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42) 
```

```python
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

```python
gds= grad_des(0.1,100)
w , b  = gds.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract scalars
def extract(val):
    if hasattr(val, 'iloc'):
        return float(val.iloc[0])
    return float(val)

w_vals = [extract(wi) for wi in w]
b_vals = [extract(bi) for bi in b]

# ── 1. Weight & Bias over Epochs ──────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(w_vals)
axes[0].set_title("Weight over Epochs")
axes[0].set_xlabel("Epoch"); axes[0].set_ylabel("Weight")

axes[1].plot(b_vals)
axes[1].set_title("Bias over Epochs")
axes[1].set_xlabel("Epoch"); axes[1].set_ylabel("Bias")

plt.tight_layout(); plt.show()

# ── 2. Actual vs Predicted ─────────────────────────────────────
final_w = extract(w[-1])
final_b = extract(b[-1])

y_pred_np = np.array([final_w * float(x_test.iloc[i, 0]) + final_b 
                       for i in range(len(x_test))])
y_test_np = np.array(y_test, dtype=float)

plt.figure(figsize=(8, 6))
plt.scatter(y_test_np, y_pred_np, s=5, alpha=0.4)
mn, mx = np.nanmin(y_test_np), np.nanmax(y_test_np)
plt.plot([mn, mx], [mn, mx], color='red', label='Ideal')
plt.xlabel("Actual"); plt.ylabel("Predicted")
plt.title("Actual vs Predicted"); plt.legend(); plt.show()

# ── 3. Residual Plot ───────────────────────────────────────────
residuals = y_test_np - y_pred_np

plt.figure(figsize=(8, 5))
plt.scatter(y_pred_np, residuals, s=5, alpha=0.4)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Predicted"); plt.ylabel("Residuals")
plt.title("Residual Plot"); plt.show()
```
