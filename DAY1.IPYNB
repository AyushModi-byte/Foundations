```python
class mat:
    def __init__(self, row, col): 
        self.row = row
        self.col = col
        # Create a 2D list filled with zeros to hold our matrix data
        self.mat = [[0 for _ in range(col)] for _ in range(row)]

    def get_mat(self):
        for i in range(self.row):
            for j in range(self.col):
                # Convert input to float (or int) so we can do math with it
                self.mat[i][j] = float(input(f"Enter value for element ({i},{j}) : "))
    
    def __add__(self, other):
        if self.row == other.row and self.col == other.col:
            c = mat(self.row, self.col)
            for i in range(self.row): 
                for j in range(self.col): 
                    c.mat[i][j] = self.mat[i][j] + other.mat[i][j]
            return c
        else:
            print("! the dimensions dont match !! ")

    def __mul__(self, other):
        if self.col == other.row: 
            # Resulting matrix has rows of the first and cols of the second
            c = mat(self.row, other.col)
            for i in range(self.row):
                for j in range(other.col): 
                    for k in range(self.col): 
                        # Dot product of row i from first and col j from second
                        c.mat[i][j] += self.mat[i][k] * other.mat[k][j]
            return c
        else:
            print("! the dimensions dont match !! ")

    def __repr__(self):
        # This allows you to easily print(matrix) and see a readable grid
        return "\n".join([str(row) for row in self.mat])
    
    def transpose(self):
        # Create a new matrix with swapped dimensions
        c = mat(self.col, self.row) 
        
        for i in range(self.row):
            for j in range(self.col):
                # Put the element at (i, j) from the original into (j, i) of the new matrix
                # Notice we use c.mat to access the underlying list!
                c.mat[j][i] = self.mat[i][j]
                
        return c
```
